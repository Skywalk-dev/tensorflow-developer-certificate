Getting information from tensors

When deeling with tensors you should be aware of the following atributes:

* Shape -> The length (number of elements) of each of the dimentions of a  tensor code tensor.shape
* Rank -> The number of tensor dimentions.A scalar has rank 0, a vector has rank1, a matrix has rank 2, a tensor has rank n. code tensor.ndim
* Actions or dimentions -> A perticular dimention of a tensor. tensor[0],tensor[:,1]
* Size -> the total number of items in the tensor code tf.size(tensor)


```python
## Create rank 4 tensors(4 dimention)

rank_4_tensor = tf.zeros(shape = (2,3,4,5))
rank_4_tensor
```

```python
#rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
## get variable attributes of our tensors (print statements)

print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimentions:", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensors:",tf.size(rank_4_tensor))
print("Total number of elements in our tensors:",tf.size(rank_4_tensor).numpy())
```

Matrix multiplication

In machine learning matrix operation is one of the most common tensor operation

There are two rules our matrix or tensors needs to fulfil if we are going to matrix multiply them.

1. The inner dimention must match.
2. The resulting matrix has the shape of the outer dimention



```python
## Matrix multiplication in tensor
tensor_2 = tf.constant([[9,12],
 [13,14]])
print(tensor)
tf.linalg.matmul(tensor,tensor_2)
```

```python
## Matrix multiplication of tensors
tensorm1 = tf.constant([[1,2,1],[0,1,0],[2,3,4]])
tensorm2 = tf.constant([[2,5],[6,7],[1,8]])
tf.matmul(tensorm1,tensorm2)
tensorm1.ndim,tensorm2.ndim
tensorm1.shape,tensorm2.shape
```

```python
tensorm1 @ tensorm2
```

```python
# Create a tensor(3,2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create another tensor of shape (3,2)
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
# Let us matrix multiply tensors of same shape

tf.matmul(X,Y)
```

```python
Y
```

```python
# Lets change the shape of Y

tf.reshape(Y,shape = (2,3))
```

```python
# Lets try to matrix multiply X with reshaped Y
X@ tf.reshape(Y, shape=(2,3))
```

```python
tf.reshape(X,shape=(2,3)) @ Y
```

```python
## using tf.matmul() for multiplication
tf.matmul(X, tf.reshape(Y, shape = (2,3)))
```

```python
# Try change the shape of X instead of Y
tf.matmul(tf.reshape(X, shape=(2,3)), Y)
```

```python
# we can do the same with transpose
## the difference between transpose and reshape is
# ## The teanspose flips the axises where as reshape shuffels the tensor
# randomly into the shape you want
X, tf.transpose(X) , tf.reshape(X , shape=(2,3))
```

```python
# So try matrix multiplication using transpose rather than reshape

tf.matmul(X, tf.transpose(Y)),
tf.matmul(tf.transpose(X),Y)
```

**Lets go through the following form of agrigation**

*   Get the minimum
*   Get the maximum
*   Get the mean of a tensor
*   Get the sum of a tensor


```python
import numpy as np
```

```python
# create a random tensor with value between 0 to 100 of size 50

E = tf.constant(np.random.randint(0,100, size = 50))
E
```

```python
tf.size, E.shape, E.ndim
```

```python
# Find the minimum of a tensor
print(tf.reduce_min(E)),
# Find the maximum of a tensor
print(tf.reduce_max(E)),
# Find the mean of a tensor
print(tf.reduce_mean(E)),
#Find the sum of a tensor
print(tf.reduce_sum(E))
```

```python
tf.reduce_mean(E)
```

```python
import tensorflow_probability as tfp
```

```python
# Finfd the varients of a tensor, we need access to tensor flow probablity
#tfp.stats.variance(E)
tf.math.reduce_variance(tf.cast(E, dtype = tf.float32))
```

```python
# find the standard deviation
tf.math.reduce_std(tf.cast(E, dtype = tf.float32))
```

Find the positional maximum and minimum


```python
# create a tensor for finding positional maximum and minimum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
np.argmax(F)
```

```python
#index of our argmax F
print(F[tf.argmax(F)]),
# Find the max value of F
print(tf.reduce_max(F))

# Check for equality
print(F[tf.argmax(F)] == tf.reduce_max(F))
```
